```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense,LSTM
from keras.layers import SimpleRNN
from tensorflow.keras.layers import Dropout
from keras.callbacks import ModelCheckpoint,EarlyStopping
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.metrics import mean_absolute_error,r2_score
```

```python
df = pd.read_csv("Kaliningrad_weather_19102020_15102011.csv",sep=';', encoding='windows-1251')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
print(f"DD: {df['DD'].unique()}")
```

```python
dictionary = {'Ветер':0, 'Штиль':1}
df['DD'] = df['DD'].map(dictionary)
```

```python
df.head()
```

```python
data1 = df[['LocalTime', 'T']]
```

```python
data1
```

```python
plt.plot(range(1,len(data1['T'].values)+1),data1['T'].values)
```

```python
min_max_scaler = MinMaxScaler()
data1 = data1['T'].values
data1 = min_max_scaler.fit_transform(data1.reshape(-1, 1))
```

```python
print(data1.shape)
data1 = data1.flatten()
print(data1.shape)
```

```python
window = 15

n_samples=data1.shape[0]-window
n_train_samples=round(0.7*n_samples)
n_val_samples=round(0.15*n_samples)
n_test_samples=n_samples-n_train_samples-n_val_samples
print('Train = ',n_train_samples)
print('Validation = ',n_val_samples)
print('Test = ',n_test_samples)
```

```python
X_train=np.zeros((n_train_samples,window))
y_train=np.zeros(n_train_samples)
```

```python
X_val=np.zeros((n_val_samples,window))
y_val=np.zeros(n_val_samples)
```

```python
X_test=np.zeros((n_test_samples,window))
y_test=np.zeros(n_test_samples)
```

```python
for i in range(n_train_samples):
  for j in range(window):
    X_train[i,j]=data1[i+j]
  y_train[i]=data1[i+window]
```

```python
for i in range(n_val_samples):
  for j in range(window):
    X_val[i,j]=data1[n_train_samples+i+j]
  y_val[i]=data1[n_train_samples+i+window]
```

```python
for i in range(n_test_samples):
  for j in range(window):
    X_test[i,j]=data1[n_train_samples+n_val_samples+i+j]
  y_test[i]=data1[n_train_samples+n_val_samples+i+window]
```

```python
X_train = np.reshape(X_train,(X_train.shape[0],X_train.shape[1],1))
X_val = np.reshape(X_val,(X_val.shape[0],X_val.shape[1],1))
X_test = np.reshape(X_test,(X_test.shape[0],X_test.shape[1],1))
```

### Данные для примера 2

```python
data2 = df.drop('LocalTime', axis=1)
```

```python
min_max_scaler = MinMaxScaler()
data2 = data2['T'].values
data2 = min_max_scaler.fit_transform(data2.reshape(-1, 1))
```

```python
print(data2.shape)
data2 = data2.flatten()
print(data2.shape)
```

```python
window_2 = 20

n_samples=data2.shape[0]-window_2
n_train_samples=round(0.8*n_samples)
n_val_samples=round(0.1*n_samples)
n_test_samples=n_samples-n_train_samples-n_val_samples
print('Train = ',n_train_samples)
print('Validation = ',n_val_samples)
print('Test = ',n_test_samples)
```

```python
X_train_2=np.zeros((n_train_samples,window_2))
y_train_2=np.zeros(n_train_samples)

X_val_2=np.zeros((n_val_samples,window_2))
y_val_2=np.zeros(n_val_samples)

X_test_2=np.zeros((n_test_samples,window_2))
y_test_2=np.zeros(n_test_samples)
```

```python
for i in range(n_train_samples):
  for j in range(window_2):
    X_train_2[i,j]=data2[i+j]
  y_train_2[i]=data2[i+window_2]

for i in range(n_val_samples):
  for j in range(window_2):
    X_val_2[i,j]=data2[n_train_samples+i+j]
  y_val_2[i]=data2[n_train_samples+i+window_2]

for i in range(n_test_samples):
  for j in range(window_2):
    X_test_2[i,j]=data2[n_train_samples+n_val_samples+i+j]
  y_test_2[i]=data2[n_train_samples+n_val_samples+i+window_2]
```

```python
X_train_2 = np.reshape(X_train_2,(X_train_2.shape[0],X_train_2.shape[1],1))
X_val_2 = np.reshape(X_val_2,(X_val_2.shape[0],X_val_2.shape[1],1))
X_test_2 = np.reshape(X_test_2,(X_test_2.shape[0],X_test_2.shape[1],1))
```

## 1. Регрессор, использующий слой RNN в качестве скрытого слоя

### Пример 1

```python
rnn_model = Sequential()
rnn_model.add(SimpleRNN(16,activation='relu',input_shape=(window,1)))
rnn_model.add(Dense(1))
```

```python
rnn_model.summary()
```

```python
rnn_model.compile(loss='mse',optimizer='adam',metrics='mae')
```

```python
early_stop = EarlyStopping(monitor='val_mae',patience=5,mode='min',verbose=1)
checkpoint = ModelCheckpoint('/content/gdrive/My Drive/StudyData/Models/ff-temperature-{epoch:02d}-{val_mae:.3f}.hdf5',
                              monitor='val_mae',verbose=1,mode='min',
                              save_best_only=True)
callbacks_list=[early_stop,checkpoint]
```

```python
rnn_model_history = rnn_model.fit(X_train,y_train,batch_size=25,epochs=40,
                                callbacks=callbacks_list,
                                validation_data=(X_val,y_val))
```

### MSE, MAE, R2

```python
y_pred = rnn_model.predict(X_test)

mse_1 = mean_squared_error(y_test, y_pred)
mae_1 = mean_absolute_error(y_test, y_pred)
r2_1 = r2_score(y_test, y_pred)

print(f'MSE: {mse_1}')
print(f'MAE: {mae_1}')
print(f'R2: {r2_1}')
```

```python
loss_function = rnn_model_history.history['loss']
val_loss_function = rnn_model_history.history['val_loss']

epochs=range(1,len(loss_function)+1)

plt.title('MSE value (Train & Val Sets)')
plt.plot(epochs,loss_function,color='green',label='Train MSE')
plt.plot(epochs,val_loss_function,color='red',label='Val MSE')
plt.xlabel('Epochs')
plt.ylabel('Mean Squared Error')
plt.legend()
plt.show()
```

```python
plt.clf()
acc_values = rnn_model_history.history['mae']
val_acc_values = rnn_model_history.history['val_mae']
plt.plot(epochs, acc_values, 'b', label='Training mae')
plt.plot(epochs, val_acc_values, 'r', label='Validation mae')
plt.title('Training and validation mae')
plt.xlabel('Epochs')
plt.ylabel('mae')
plt.legend()
plt.show()
```

```python
y_pred = rnn_model.predict(X_test)
y_pred_inv = min_max_scaler.inverse_transform(y_pred)
y_test_inv = min_max_scaler.inverse_transform(y_test.reshape(-1,1))

print('MAE = ',round(mean_absolute_error(y_true=y_test_inv,y_pred=y_pred_inv),3),' K degrees')
print('R2-score = ',round(r2_score(y_test_inv,y_pred_inv),3))
```

```python
plt.plot(range(1,len(y_test_inv)+1),y_test_inv)
plt.plot(range(1,len(y_pred_inv)+1),y_pred_inv)
```

### Сравнение

```python
models = ['Model 1', 'Model 2']

mse_val = [mse_1, mse_2]
mae_val = [mae_1, mae_2]
r2_val = [r2_1, r2_2]
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(10,8))

# Диаграмма MSE для всех рассмотренных моделей
axes[0].bar(models, mse_val, color='red')
axes[0].set_ylabel('MSE')
axes[0].set_title('MSE для моделей')

# Диаграмма MAE для всех рассмотренных моделей
axes[1].bar(models, mae_val,color='green')
axes[1].set_ylabel('MAE')
axes[1].set_title('MAE для моделей')

# Диаграмма R2 для всех рассмотренных моделей
axes[2].bar(models, r2_val,color='violet')
axes[2].set_ylabel('R2')
axes[2].set_title('R2 для моделей')

plt.tight_layout()
plt.show()
```

## Регрессор – стек двух слоёв с LSTM (или GRU)

### Пример 1

```python
model = Sequential()
model.add(LSTM(16, activation='relu', input_shape=(window,1), return_sequences=True))
model.add(LSTM(8, activation='relu'))
model.add(Dense(1))
```

```python
model.summary()
```

```python
model.compile(loss='mse',optimizer='adam',metrics='mae')
```

```python
model_history = model.fit(X_train, y_train, batch_size=40, epochs=30,validation_data=(X_val,y_val))
```

### MSE, MAE, R2

```python
y_pred = model.predict(X_test)

mse_1 = mean_squared_error(y_test, y_pred)
mae_1 = mean_absolute_error(y_test, y_pred)
r2_1 = r2_score(y_test, y_pred)

print(f'MSE: {mse_1}')
print(f'MAE: {mae_1}')
print(f'R2: {r2_1}')
```

```python
loss_values = model_history.history['loss']
val_loss_values = model_history.history['val_loss']

epochs = range(1, len(loss_values) + 1)
plt.plot(epochs, loss_values, 'b', label='Train MSE')
plt.plot(epochs, val_loss_values, 'r', label='Validation MSE')
plt.title('MSE value (Train&Val sets)')
plt.xlabel('Epochs')
plt.ylabel('MSE')
plt.legend()
plt.show()
```

```python
plt.clf()

acc_values = model_history.history['mae']
val_acc_values = model_history.history['val_mae']

plt.plot(epochs, acc_values, 'b', label='Training mae')
plt.plot(epochs, val_acc_values, 'r', label='Validation mae')
plt.title('Training and validation mae')
plt.xlabel('Epochs')
plt.ylabel('mae')
plt.legend()
plt.show()
```
